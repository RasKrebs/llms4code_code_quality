### Library Import 

```python
# Author produced code
from baseline import PrincipalComponentAnalysis as Baseline_PCA

# Sklearn version - industry standard
from sklearn.decomposition import PCA as Sklearn_PCA

# GPT 3.5
from gpt_35 import PrincipalComponentAnalysis as GPT35_PCA


# General imports
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import numpy as np
import seaborn as sns
```

```python
# Load the data
data = np.loadtxt('../data/bodyfat.txt', skiprows=11)

# Normalize the data
scaler = StandardScaler()
data = scaler.fit_transform(data)

# Print the first 2 rows
print('First data row:\n', data[1])

# Print shape
print('\nData shape: ', data.shape)
```

### Applying
Creating PCA and returning 2 components

```python
#### BASELINE
base_pca = Baseline_PCA()
base_components = base_pca.fit_transform(data, n_components=2)

#### SKLEARN
sklearn_pca = Sklearn_PCA(n_components=2)
sklearn_components = sklearn_pca.fit_transform(data)

#### GPT 3.5
gpt35_pca = GPT35_PCA()
gpt35_componenets = gpt35_pca.fit_transform(data, n_components=2)
```

```python
sns.scatterplot(x=base_components[:, 0], y=base_components[:, 1], label='Baseline')
sns.scatterplot(x=sklearn_components[:, 0], y=sklearn_components[:, 1], label='Sklearn')
sns.scatterplot(x=gpt35_componenets[:, 0], y=gpt35_componenets[:, 1], label='GPT 3.5')
```
